```python
import sklearn

discours = sklearn.datasets.load_files('/content/drive/MyDrive/dataset_ITOR_Boostrap')
```

```python
import os
import re

dico = {}
liste = []
titre = []
i=0
for discour in discours.filenames:
  dico[os.path.basename(discour)] = re.sub('[!,*)@#%(&$_?.^]', '', discours.data[i].decode('utf8', "replace").replace('\n',' '))
  liste.append(re.sub('[!,*)@#%(&$_?.^]', '', discours.data[i].decode('utf8', "replace").replace('\n',' ')))
  titre.append(os.path.basename(discour))
  i += 1
dico
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import numpy as np
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop

Vectorizer = TfidfVectorizer(stop_words = fr_stop)
X = Vectorizer.fit_transform(liste)

true_k = 2
model = KMeans(n_clusters=true_k,max_iter=100, init='k-means++', n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = Vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
```

```python
import math
import re

idfDico = {}
wordAlreadyCounted = []
string_check= re.compile('[@_!#$%^&*()<>?/\|}{~:]') 
i = 0

for doc in liste:
  sentence = doc.split()
  dico_doc = {}
  for word in sentence:
    if((word not in wordAlreadyCounted) and string_check.search(word) == None):
      result = re.findall(word, doc, re.IGNORECASE)
      dico_doc[word] = len(result) / len(sentence)
      wordAlreadyCounted.append(word)

  wordAlreadyCounted.clear()
  idfDico[titre[i]] = dico_doc
  i += 1

print(idfDico)
```
