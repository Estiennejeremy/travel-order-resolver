## Création des datasets

```python
class Sentence:
    def __init__(self, text, president):
        self.text = text
        self.president = president

sentences = [];
for president, data in presidents.items():
    for sent in data["sentences"]:
        sentences.append(Sentence(sent, president))
print("Nombre de phrases", len(sentences))
print(sentences[0].president, sentences[0].text)
```

## Construction du model

```python
from sklearn.model_selection import train_test_split

training, test = train_test_split(sentences, test_size=0.33, random_state=42)
```

```python
class SentenceContainer:
    def __init__(self, reviews):
        self.reviews = reviews
        
    def get_text(self):
        return [x.text for x in self.reviews]
    
    def get_president(self):
        return [x.president for x in self.reviews]
    
train_container = SentenceContainer(training)
test_container = SentenceContainer(test)
```

```python
train_x = train_container.get_text()
train_y = train_container.get_president()

test_x = test_container.get_text()
test_y = test_container.get_president()

print(train_y[:10])
```

### Vectorisation

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

vectorizer = TfidfVectorizer()
train_x_vectors = vectorizer.fit_transform(train_x)
test_x_vectors = vectorizer.transform(test_x)

print(train_x[0])
print(train_x_vectors[0].toarray())
```

### Classification & prédiciton

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(train_x_vectors, train_y)

print(clf.predict(test_x_vectors[1]), test_y[1])
```

## Evaluation

```python
# Mean Accuracy
print(clf.score(test_x_vectors, test_y))
```
