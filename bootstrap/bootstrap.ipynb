```python
# https://towardsdatascience.com/natural-language-processing-feature-engineering-using-tf-idf-e8b9d00e7e76

import re
import os

directory = "./discours/tous"

all_content = ""; # tous le contenut de tout les fichiers
for filename in os.listdir(directory): # pour chaque fichier dans le répertoire
    if filename.endswith(".txt"): # si le fichier a pour extension .txt
        f = open(os.path.join(directory, filename), "r") # ouverture du fichier en lecture
        content = f.read() # récupération du contenu
        all_content = all_content + " " + content
    else:
        continue
        
        
all_words = re.split('; |, |\*|\n| |\.|:|\'|"',all_content); # liste de tous les mots
all_unique_words = set(all_words) # list de tous les mots sans doublon

num_of_words = dict.fromkeys(all_unique_words, 0); # list de nombre de mots

for word in all_words: # pour tous les mots on calcule combien il y en a
    num_of_words[word] += 1
    
print(num_of_words) 

```

```python
import os

directory = "./discours/tous"

presidents = {}; #liste de tous les présidents ansi que les fichiers associés
for filename in os.listdir(directory): # pour chaque fichier dans le répertoire
    if filename.endswith(".txt"): # si le fichier a pour extension .txt
        president = filename.split("_")[0] + "_" + filename.split("_")[1];
        if president in presidents:
            presidents[president]["fichiers"].append(filename);
        else:
            presidents[president] = {"fichiers": [filename] };
    else:
        continue

print(presidents)
        
```

```python
'''
pip install -U pip setuptools wheel
pip install -U spacy
python -m spacy download fr_core_news_sm
'''
# Chargement de lib (fr)
import spacy
nlp = spacy.load('fr_core_news_sm')
```

```python
f = open(os.path.join(directory, "Alain_Madelin_702.txt"), "r") # ouverture du fichier en lecture
content = f.read() # récupération du contenu
```

```python
## Tokenisation par mots
docx = nlp(content)
for token in docx:
    print(token.text)
```

```python
## Tokenisation par mots
for num,sent in enumerate(docx.sents,start=1):
    print(f"{num}:{sent}")
```

```python
# Lemmatisation & Stemming
for token in docx:
    print((token.text,token.lemma_))
```

```python
# Etiquetage
for token in docx:
    print((token.text,token.tag_))
```

```python
spacy.explain('ADP')
```

```python
from spacy import displacy

index = 0; 
for num,sent in enumerate(docx.sents,start=1):
    if(index == 1):
        displacy.render(sent,style='dep',jupyter=True)
        break
    index +=1
```

```python
# Reconnaissance d’entités nommées
for entity in docx.ents:
    print((entity.text,entity.label_))
```

```python
# Vectorisation
for word in docx:
    print(word.text,word.vector)
```

```python
# Détection des stops words

for token in docx:
    print(token.text,token.is_stop)
```

```python
# Suppression des stops words
non_stopwords = [token.text for token in docx if token.is_stop != True]
print(non_stopwords)
```
