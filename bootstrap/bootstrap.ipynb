```python
# https://towardsdatascience.com/natural-language-processing-feature-engineering-using-tf-idf-e8b9d00e7e76

import re
import os

directory = "./discours/tous"

all_content = ""; # tous le contenut de tout les fichiers
for filename in os.listdir(directory): # pour chaque fichier dans le répertoire
    if filename.endswith(".txt"): # si le fichier a pour extension .txt
        f = open(os.path.join(directory, filename), "r") # ouverture du fichier en lecture
        content = f.read() # récupération du contenu
        all_content = all_content + " " + content
    else:
        continue
        
        
all_words = re.split('; |, |\*|\n| |\.|:|\'|"',all_content); # liste de tout les mots
all_unique_words = set(all_words) # list de tous les mots sans doublon

num_of_words = dict.fromkeys(all_unique_words, 0); # list de nombre de mots

for word in all_words: # pour tous les mots on calcule combien il y en a
    num_of_words[word] += 1
    
print(num_of_words) 

```
